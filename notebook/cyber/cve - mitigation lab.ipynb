```python
# inspect our mitigation code

import os, pickle
from collections import defaultdict, Counter

entset = pickle.load(open(os.path.join('./basic/', 'entset.pkl'), 'rb'))
factset = pickle.load(open(os.path.join('./basic/', 'factset.pkl'), 'rb'))

cve_miti_map = defaultdict(set)
miti_cve_map = defaultdict(set)
for rel, facts in factset.items():
    if rel not in ['MITI:mitigates:CVE']:
        continue
    for h, r, t in facts:
        assert r==rel
        assert h in entset['mitigation']
        assert t in entset['cve-id']
        cve_miti_map[t].add(h)
        miti_cve_map[h].add(t)
        
miti2cve_fq = []
for k,v in miti_cve_map.items():
    miti2cve_fq.append(len(v))
count_dict = Counter(miti2cve_fq)
cum = 0
for k in sorted(count_dict.keys(), reverse=True):
    cur_ratio = count_dict[k]*k/sum(miti2cve_fq) 
    cum += cur_ratio
    print('miti- %d cve: frequency %d, ratio %.4f, cumulative ratio %.4f' % 
          (k, count_dict[k], cur_ratio, cum))
```
