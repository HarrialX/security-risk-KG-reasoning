```python
# check each disease has how many treatment drugs

import os, sys, pickle
from collections import defaultdict

sys.path.append(os.path.abspath('../..'))

import genkg.cyberkg_utils as cyber

kg_path = '/data/zhaohan/adv-reasoning/save/data/medkg'

id_factset = pickle.load(open(os.path.join(kg_path, 'id_factset.pkl'), 'rb'))
id_entset = pickle.load(open(os.path.join(kg_path, 'id_entset.pkl'), 'rb'))
id2ent = pickle.load(open(os.path.join(kg_path, 'id2ent.pkl'), 'rb'))   
rel2id = pickle.load(open(os.path.join(kg_path, 'rel2id.pkl'), 'rb')) 
id2rel = pickle.load(open(os.path.join(kg_path, 'id2rel.pkl'), 'rb')) 

rid_1 = rel2id['reverse:positive:Compound:Disease']
rid_2 = rel2id['neutral:Disease:Symptom']

fact_dict = cyber.gen_factdict(kg_path)

evi_fq = defaultdict(int)

for eid in id_entset['Disease']:
    if rid_1 not in fact_dict[eid]:
        continue
    if rid_2 not in fact_dict[eid]:
        continue

    drug_ids = fact_dict[eid][rid_1]
    for drug_id in drug_ids:
        assert drug_id in id_entset['Compound']
        
    evi_ids = fact_dict[eid][rid_2]
    for e_id in evi_ids:
        assert e_id in id_entset['Symptom']
        evi_fq[e_id] += 1
        
    print(eid, id2ent[eid], len(drug_ids), len(evi_ids))
```
